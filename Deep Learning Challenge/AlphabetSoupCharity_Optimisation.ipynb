## AlphabetSoupCharity_Optimisation: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import sklearn as skl

#  Import and read the charity_data.csv.
import pandas as pd 
Application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
Application_df.head(10)
```

```python
# Drop the non-beneficial ID columns, 'EIN','NAME', and 'INCOME_AMT'

Application_df = Application_df.drop(columns=['EIN','NAME','SPECIAL_CONSIDERATIONS','INCOME_AMT'])
Application_df.head()
```

```python
# Determine the number of unique values in each column.

unique_counts = Application_df.nunique()
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
Application_type_counts =  Application_df['APPLICATION_TYPE'].value_counts()
print(Application_type_counts)
```

```python
# Choose a cutoff value. cutoff incresed from 500 to 1000
cutoff_value = 1000

# Create a list of application types to be replaced
Application_types_to_replace = Application_type_counts[Application_type_counts < cutoff_value].index.tolist()

# Replace in dataframe
for app in Application_types_to_replace:
    Application_df['APPLICATION_TYPE'] = Application_df['APPLICATION_TYPE'].replace(app, "Other")

# Check to make sure binning was successful
Application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning

classification_counts =  Application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
print(classification_counts[classification_counts > 1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1000
# Create a list of classification to be replaced
classifications_to_replace = classification_counts[classification_counts < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    Application_df['CLASSIFICATION'] = Application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
Application_df['CLASSIFICATION'].value_counts()
```

```python
# Determine categorcal variables in dataframe
Application_df.dtypes
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
Application_numeric = pd.get_dummies(Application_df)


Application_numeric.head()
```

```python
# Split our preprocessed data into our features and target arrays
y= Application_numeric["IS_SUCCESSFUL"].values
X= Application_numeric.drop(columns="IS_SUCCESSFUL").values

# Split the preprocessed data into a training and testing dataset. random_state  reduced
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=58)  
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
print(X_train_scaled.shape)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=29))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=50, activation="relu"))

# third hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation="relu"))

# fourth hidden layer
nn.add(tf.keras.layers.Dense(units=40, activation="relu"))

# fifth hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
# nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
# Define the optimizer with a custom learning rate
custom_optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001)

# Compile the model with the custom optimizer
nn.compile(loss="binary_crossentropy", optimizer=custom_optimizer, metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=40,  batch_size=29)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity.h5")
```
